```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Base_Socio_Formador_Con_Nulos.csv', encoding = "latin")
data.head(50)
```

```python
#Realizamos una copia del dataframe
data2 = data.copy()
```

```python
data2['precio']= data2['precio'].replace({'\$': '', ',': ''}, regex=True)
data2['precio']=pd.to_numeric(data2['precio'], errors = 'coerce')
data2= data2.dropna(subset=['precio'])
data2['precio']=data2['precio'].astype(int)
print(data2['precio'].head(5))
```

```python
#Converir los datos nulos a promedio en la columna precio
data2["precio"]=data2["precio"].fillna(round(data2["precio"].mean(),1))
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
'''
El método "bfill" (backfill) toma el valor no nulo más cercano posterior a un valor faltante y lo usa para rellenar.
Esto es útil si quieres asegurar que las fechas no queden en blanco y que se respete el flujo temporal hacia adelante.
'''
data2["fecha"] = data2["fecha"].fillna( method = "bfill")
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
'''
Si los datos de año_hecho están relacionados con eventos consecutivos o hechos históricos que ocurrieron de manera continua
(por ejemplo, datos de un censo o registros anuales), tiene sentido que los años faltantes se llenen con el año del evento más reciente.
'''
data2["año_hecho"] = data2["año_hecho"].fillna(method = "ffill")
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
'''
Si los datos de año_renovado se refieren a la renovación de algo (contratos, licencias, productos, etc.),
puede ser común que varios registros en una misma fila de tiempo tengan el mismo año de renovación.
'''
data2["año_renovado"] = data2["año_renovado"].fillna(method = "ffill")
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2["metros_cuadrados"]=data2["metros_cuadrados"].fillna(round(data2["metros_cuadrados"].mean(),1))
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
'''
Si pisos se refiere a niveles de edificios u otro tipo de estructura, es posible que un grupo de
registros esté relacionado con un mismo piso. Al usar ffill, los valores faltantes se completan con el piso anterior
'''
data2["pisos"] = data2["pisos"].fillna(method = "ffill")
```

```python
'''
Coherencia en el número de habitaciones: Si los datos representan propiedades o espacios (como casas o departamentos) y
algunas filas no tienen el número de habitaciones registrado,
bfill es útil para asegurarte de que todos los registros tengan un valor sin dejar espacios vacíos.
'''
data2["habitaciones"] = data2["habitaciones"].fillna( method = "bfill")
```

```python
'''
Al aplicar el método fillna() con el promedio redondeado para la columna baños, estás sustituyendo
los valores faltantes con el promedio de los baños existentes en el conjunto de datos.
'''
data2["baños"]=data2["baños"].fillna(round(data2["baños"].mean(),1))
```

```python
'''
Si tus datos contienen información relacionada con direcciones (por ejemplo, propiedades,
 ubicaciones de clientes, etc.), puede ser lógico rellenar los
valores nulos en la columna calle con el nombre de la siguiente calle registrada
'''
data2["calle"] = data2["calle"].fillna( method = "bfill")
```

```python
'''
Si los registros en tu DataFrame están agrupados geográficamente (por ejemplo, por áreas, vecindarios, o
bloques que comparten el mismo código postal), el uso de ffill tiene sentido
'''
data2["codigo_zip"] = data2["codigo_zip"].fillna(method = "ffill")
```

```python
'''
Si los datos representan ubicaciones (por ejemplo, propiedades, clientes, o ventas) y están organizados de tal manera
que los registros de un mismo estado están agrupados, usar ffill tiene sentido.
'''
data2["estado"] = data2["estado"].fillna(method = "ffill")
```

```python
'''
Si tus datos contienen información de ubicaciones, como ciudades
donde están situadas propiedades o clientes, usar bfill puede ser útil.
'''
data2["ciudad"] = data2["ciudad"].fillna( method = "bfill")
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```
