```python
#Creo 2 dataFrames para poder procesesar los outliers
cuantitativas = data.iloc[:, 1:24]
cualitativas = data.iloc[:,[0]]
#Dividir el dataframe por el tipo de valor por ejemplo de la columna 1 a la 23 son valores flotantes
#Y en el segundo esta metiendo solo las que son de tipo objeto y en este caso es solo la 0
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atpicos del Dataframe")
plt.show() #
```

##PROCEDIMIENTO DESVIACION ESTANDAR PARA ELIMINAR OUTLIERS EN DATAFRAME

```python
#Mtodo aplicando desviacin estandar. Encuentro los valores extremos
y = cuantitativas
limite_superior = y.mean() + 3*y.std()
limite_inferior = y.mean() - 3*y.std()
print("Limite superior permitido", limite_superior)
print("Limite inferior permitido", limite_inferior)
#Estos representan los lmites que se tienen en la campana de gauss
```

```python
#Verificamos outliers
data2 = cuantitativas[(y >= limite_superior) | (y <= limite_inferior)]
data2
#Tienen que estar afuera de la campana
```

```python
data3 = cuantitativas[(y <= limite_superior) & (y >= limite_inferior)]
data3
```

```python
#Corroboramos dataframes valores nulos del dataframe4 El metodo de desviacion detecta menos outliers
#Con el metodo rangointercualtilicoy el digrama de caja y bigote usa ese metodo de rangointercualitico
valores_nulos = data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores tipicos  (nulos deldataframe con "mean")
#Realizamos una copia del dataframe
data_clean = data3.copy()
data_clean = data_clean.fillna(round(y.mean(),1))
data_clean
```

```python
Datos_limpios_2 = pd.concat([cualitativas, data_clean], axis=1)
Datos_limpios_2
```

##Mtodo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75

```python
y = cuantitativas

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data3_iqr = cuantitativas[(y <= Limite_Superior_iqr) & (y >= Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe limpio
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
#Con este metodo detectamos mas valores atipicos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr = data3_iqr.copy()
data4_iqr = data4_iqr.fillna(round(y.mean(),1))
data4_iqr
```

```python
Datos_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```
