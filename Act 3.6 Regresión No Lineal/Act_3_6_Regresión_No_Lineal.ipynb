```python
#Cargamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("Datos_limpios (4).csv")
```

```python
#Eliminar un signo de una columna
df['host_is_superhost']=df['host_is_superhost'].str.replace('t', '1')
df['host_is_superhost']=df['host_is_superhost'].str.replace('f', '0')
df['host_is_superhost']=df['host_is_superhost'].str.replace('Desconocido', '0')
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df['host_is_superhost']= df['host_is_superhost'].astype(float)
```

```python
#Declaramos las variables dependientes e independientes para la regresión no Lineal
vars_Indep = df[['host_acceptance_rate']]
var_Dep = df['host_response_rate']
```

```python
x = vars_Indep
y = var_Dep
```

```python
def func1 (x,a,b,c):
  return a*x**2 + b*x + c
```

```python
#Ajustamos los parámetros de la función curve_fit
parametros1, covs = curve_fit(func1, df['host_acceptance_rate'], df['host_response_rate'])
```

```python
#Obtenemos los coeficientes del modelo de regresión no lineal
parametros1
```

```python
#Preducciones
y_pred1 = func1(x, *parametros1)
```

```python
#Calculamos las predicciones y reestauramos el vector de predicciones
y_pred1
```

```python
#Calculamos el coeficiente de determinación del modelo
R2_Modelo1 = r2_score(y, y_pred1)
R2_Modelo1
```

```python
#Calculamos el coeficiente de correlacion del modelo
R = np.sqrt(R2_Modelo1)
R
```

```python
#Declaramos las variables dependientes e independientes para la regresión no Lineal
vars_Indep = df[['host_acceptance_rate']]
var_Dep = df['host_response_rate']
```

```python
x = vars_Indep
y = var_Dep
```

```python
def func1(x, a, b, c):
  return (a*x**2 + b)/ c*x
```

```python
#Ajustamos los parámetros de la función curve_fit
parametros1, covs = curve_fit(func1, df['host_acceptance_rate'], df['host_response_rate'])
```

```python
#Obtenemos los coeficientes del modelo de regresión no lineal
parametros1
```

```python
#Preducciones
y_pred1 = func1(x, * parametros1)
```

```python
#Calculamos las predicciones y reestauramos el vector de predicciones
y_pred1
```

```python
#Calculamos el coeficiente de determinación del modelo
R2_Modelo1 = abs(r2_score(y, y_pred1))
R2_Modelo1
```

```python
#Calculamos el coeficiente de correlacion del modelo
R = np.sqrt(R2_Modelo1)
R
```

```python
#Graficamos las predicciones y los datos originales para realizar la comparación
plt.plot(x,y, 'bo', label="y-original")
plt.plot(x, y_pred1, label="y-predecida con y= (a*x**2 + b)/ c*x")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
