```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Hacer el archivo de excel en un dataframe
df = pd.read_excel('Inmuebles.xlsx')
```

```python
df.info()
```

 a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”

```python
#Elimimnar la priemera fila
df = pd.read_excel('Inmuebles.xlsx', skiprows=1)
```

```python
#Eliminar columnas inesearias
#axis = 1 para columnas
#axis = 2 para filas
df1 = df.drop (['Referencia'], axis=1)
df1
```

b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

```python
filtro1 = df1[(df1["Fecha_Alta"] == "2007-04-29 00:00:00")|(df1["Fecha_Alta"] == "2004-01-02 00:00:00")]

filtro1
```

c) Todos los registros de “ Tipo: parking, local, industrial y piso”

```python
filtro2 = df1[(df1["Tipo"] == "Local")|(df1["Tipo"] == "Industrial")|(df1["Tipo"] == "Industrial")| (df1["Tipo"] == "Piso")]

filtro2
```

d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

```python
filtro3= df1[(df1["Provincia"] == "Girona") & (df1["Superficie"] > 100)]

filtro3
```

e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”

```python
filtro4= df1[(df1["Precio Venta"] > 1000000)|(df1["Vendedor"] == "Carmen" ) | (df1["Vendedor"] == "Joaquin")]

filtro4
```

f) Todos los registros que no sean de “Provincia : Barcelona“

```python
filtro5 = df[~(df["Provincia"] == "Barcelona")]#QUIERO QUE NO PONGAS ESTO
filtro5
```

g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtro6 = df[~(df["Operación"] == "Alquiler") & ~ (df["Tipo"] == "Oficina")] #QUIERO QUE NO PONGAS ESTO
filtro6
```

```python
filtro6.to_csv("Entregable6.csv")
```
