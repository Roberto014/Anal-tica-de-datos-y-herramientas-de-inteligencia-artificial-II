```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Hacer el archivo de excel en un dataframe
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
df.head()
```

filtro 1 por objeto

```python
filtro1 = df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

filtro por filas

```python
Filtro2 = df.iloc[4:1424,:]
Filtro2
```

filtro 3 por filas no consecutivo

```python
filtro3 = df.iloc[[9, 20, 40,100,200], : ]
filtro3
```

Filtro 4 columnas consecutivas

```python
filtro4 = df.iloc[: , :4]  #Columnas consecutivas
filtro4
```

filtro 5

```python
filtro5 = df.iloc[3:5, [1,3,5]] #Columnas no consecutivas
filtro5
```

Filtros basicos con conectores logicos

```python
filtro9 = df[(df["PRECIO_UNITARIO"] > 0) & (df["CVE_DOC"] == "F37042")]
filtro9
```

Filtro O

```python
#Filtro O
filtro10 = df[(df["PRECIO_UNITARIO"] > 10000) | (df["CVE_DOC"] == "F37042")]
filtro10
```

```python
filtro11 = df[~(df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]#QUIERO QUE NO PONGAS ESTO
filtro11
```
