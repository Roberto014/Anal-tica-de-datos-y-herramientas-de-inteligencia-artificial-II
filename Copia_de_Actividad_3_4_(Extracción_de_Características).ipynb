```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
Data = pd.read_csv("Datos_limpios (4).csv")
```

```python
Data.info()
```

```python
#Obtenga un análisis univariado de las variables categóricas
freq_tbl(Data)
```

```python
#Obtengo un análisis univariado de una variable categorica en especifico
table1 = freq_tbl(Data['last_scraped'])
table1
```

```python
table2 = table1.drop(['percentage','cumulative_perc'], axis=1)
table2
```

```python
#Ajusto el indice de mi dataframe
Filtro_index=table2.set_index('last_scraped')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=0.8, figsize=(8,3))
plt.title('last_scrap')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')
```
